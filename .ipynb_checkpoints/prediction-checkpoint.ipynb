### Polynomial determining

```python
degrees = 10
X = df.drop(columns=['bmi', 'heart_disease', 'stroke', 'hypertension', 'gender', 'id', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'])
y = df['bmi']

train_error = []
test_error = []
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.7)

for degree in range(degrees):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_train, y_train)
    train_error.append(mean_squared_error(y_train, model.predict(X_train)))
    test_error.append(mean_squared_error(y_test, model.predict(X_test)))

plt.plot(np.arange(degrees), train_error, color='green', label='train')
plt.plot(np.arange(degrees), test_error, color='red', label='test')
plt.ylabel('mean squared error')
plt.xlabel('degree')
plt.legend(loc='upper left')
```

It would seem that degree 4 or 5 would be the optimal point before bouncing back up, but the improvements are so low I will just be using 2nd degree.

```python
degree = 2
rcv = RidgeCV(cv=10, alphas = [1e-8, 1e-4, 1e-2, 1e-1, 1, 10, 20, 40, 80, 200, 750, 1000, 5000, 10000])

model = make_pipeline(PolynomialFeatures(degree), rcv)
model.fit(X_train, y_train)

print("alpha: ", model.steps[1][1].alpha_, f"   Adjusted R^2: {1 - (1-model.score(X,y))*(len(y)-1)/(len(y)-X.shape[1]-1)}","    R^2:", model.score(X_test, y_test))
```

It would seem that the MSE of my training and testin are rather close, and the best alpha seems to be 1000.

R^2 is the square error of which the predicted model differs from the actual data, and the adjusted R^2 is the version of which that takes into account how many predictors are present in the data.

### Bias and Variance

```python
plt.plot(X_train['age'], y_train, '.', X_train['age'], model.predict(X_train), '*')
plt.xlabel('age')
plt.ylabel('bmi')
plt.legend(['actual', 'predicted'])
```

```python
plt.plot(X_train['avg_glucose_level'], y_train, '.', X_train['avg_glucose_level'], model.predict(X_train), '*')
plt.xlabel('avg glucose lvl')
plt.ylabel('bmi')
plt.legend(['actual', 'predicted'])
```

From the predict data graphed around the average glucose level and age, we can see that the variance is not too high. The prediction follows a general trend with some extending out a bit further, but nothing too extreme, therefore no overfitting, and the bias isn't too sever either, as the data is not underfitted.
